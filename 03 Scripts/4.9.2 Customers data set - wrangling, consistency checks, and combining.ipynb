# 03. Data consistency checks - df_cust

## 01. Mixed-type values

```python
# Check for mixed-type data
for col in df_cust.columns.tolist():
    weird = (df_cust[[col]].applymap(type) != df_cust[[col]].iloc[0].apply(type)).any(axis = 1)
    if len (df_cust[weird]) > 0:
        print(col)
```

```python
# Change data type of first_name column to string
df_cust['first_name'] = df_cust['first_name'].astype('str')
```

# 04. Combining data sets

```python
# Look for key columns - df_cust info
df_cust.info()
```

```python
# Look for key columns - df_ords_prods info
df_ords_prods.info()
```

```python
# Change data type of user_id column to string

df_ords_prods['user_id'] = df_ords_prods['user_id'].astype('str')
```

```python
# Check output 
df_ords_prods['user_id'].dtype
```

```python
# Drop _merge column from df_ords_prods
df_ords_prods = df_ords_prods.drop(columns = ['_merge'])
```

```python
# Check output
df_ords_prods.columns
```

```python
# Merge data sets with user_id as key column
df_merged = df_ords_prods.merge(df_cust, on = ['user_id'], indicator = True)
```
