# 03. If-statements with user-defined functions

```python
# Create subset of first million rows
df = ords_prods_merged[:1000000]
```

```python
# Check output
df.head()
```

```python
# Check shape of df
df.shape
```

```python
# Define a function
def price_label(row):
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High-range product'
    else: return 'Not enough data'
```

```python
# Apply the function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Check frequencies of new column
df['price_range'].value_counts(dropna = False)
```

```python
# Check most expensive product
df['prices'].max()
```

# 04. If-statements with the loc() function

```python
# Create high-range product flag on df
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High range product'
```

```python
# Create mid-range product flag on df
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid range product'
```

```python
# Create low-range product flag on df
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low range product'
```

```python
# Check frequency of new column in df
df['price_range_loc'].value_counts(dropna = False)
```

```python
# Create high-range product flag on ords_prods_merged
ords_prods_merged.loc[ords_prods_merged['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
# Create mid-range product flag on ords_prods_merged
ords_prods_merged.loc[(ords_prods_merged['prices'] <= 15) & (ords_prods_merged['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
# Create low-range product flag on ords_prods_merged
ords_prods_merged.loc[ords_prods_merged['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
# Check frequency of new column in ords_prods_merged
ords_prods_merged['price_range_loc'].value_counts(dropna = False)
```

# 05. If-statements with for-loops

```python
# Simple example of for-loop
for x in range(30, 45):
    print('My age is %d' % (x))
```

```python
# Check frequency of orders_day_of_week
ords_prods_merged['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Create result list of results from for-loop classifying busiest day, least busy, or regularly busy
result = []
for value in ords_prods_merged['orders_day_of_week']:
    if value == 0:
        result.append('Busiest day')
    elif value == 4:
        result.append('Least busy')
    else:
        result.append('Regularly busy')
```

```python
# Check output
result
```

```python
# Add new column to ords_prods_merged equal to result
ords_prods_merged['busiest_day'] = result
```

```python
# Check frequency of busiest_day to compare with previous orders_day_of_week
ords_prods_merged['busiest_day'].value_counts(dropna = False)
```

# 06. Deriving new variables (Task 4.7)

```python
# Check frequency of orders_day_of_week
ords_prods_merged['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Create new_results list of results from for_loop classifying busiest days, least busy days, or regulary busy
new_results = []
for value in ords_prods_merged['orders_day_of_week']:
    if value == 0:
        new_results.append('Busiest days')
    elif value == 1:
        new_results.append('Busiest days')
    elif value == 4:
        new_results.append('Least busy days')
    elif value == 3:
        new_results.append('Least busy days')
    else:
        new_results.append('Regularly busy')
```

```python
# Check output
new_results
```

```python
# Add new column to ords_prods_merged equal to new_results
ords_prods_merged['busiest_days'] = new_results
```
